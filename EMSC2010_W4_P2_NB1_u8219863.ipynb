##Importing Python Packages

```python
import pandas as pd #used to read and clean data
import matplotlib.pyplot as plt #used for data plotting
```

##Getting Github to read my data

```python
file_path = "/content/co2_mm_mlo.csv"
df = pd.read_csv("/content/co2_mm_mlo.csv", skiprows=40) #Telling it to skip the first 40 rows, with random words
print(df.head())
```

##Plotting the Data

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['decimal date'],df['average'],'k', color='red') #plot time vs. Z on ax as a black ('k') line
ax.set_xlabel('Year') #text for the x-axis label
ax.set_ylabel('Averaga CO\u2082 level (ppm)') #text for the y-axis label
ax.set_xlim([1980,2030]) #interval for the x-axis
ax.set_ylim([320,480]) #interval for the x-axis
ax.minorticks_on() #show the minor ticks
plt.title('The increase in CO\u2082 levels from 1980 to 2026')
```
